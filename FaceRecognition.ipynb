```python
#LBPH MODEL
import cv2
from PIL import Image
import numpy as np
from os import listdir
from os.path import isfile,join

dataPath = 'path of the folder'

onlyFiles = [f for f in listdir(dataPath) if isfile(join (dataPath, f))]#picking up all the files or photos from ur directory

trainingData, Lables = [], [] #two empty lists

for i, files in enumerate(onlyFiles):
    img_path = dataPath + onlyFiles[i] #paths for images,,,C://Users//Harry//Documents//MLOPS//harFace//image 1.jpg
                                                         #,C://Users//Harry//Documents//MLOPS//harFace//image 1.jpg, etc.
    imgs = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    
    trainingData.append(np.asarray(imgs, dtype=np.uint8))#converting the list and storing value init as an array
    Lables.append(i)#for retaing count value
    
Lables = np.asarray(Lables, dtype=np.int32)#convert it to array
model=cv2.face_LBPHFaceRecognizer.create()#model creation

model.train(np.asarray(trainingData), np.asarray(Lables))#model trining        

print("Face ID SAVED!!")
```

```python
import cv2
import numpy as np
import os
import IPython.display as ipd
import webbrowser as wb

face_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
#for showing cropped image being clicked live

def face_detector(img, size=0.5):

    # Convert image to grayscale
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces = face_classifier.detectMultiScale(gray, 1.3, 5)
    if faces is ():
        return img, []

    for (x,y,w,h) in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,255),2)
        roi = img[y:y+h, x:x+w]
        roi = cv2.resize(roi, (200, 200))
    return img, roi



cap = cv2.VideoCapture(0)

while True:

    ret, frame = cap.read()

    image, face = face_detector(frame)#calling function here

    try:
        face = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)

        result = model.predict(face)#fitting into model
        print(result)
        if result[1] < 500:
            cnf = int( 100 * (1 - (result[1])/400) )
            #display_string = str(cnf) + '% Confident it is User'

        #cv2.putText(image, display_string, (100, 120), cv2.FONT_HERSHEY_COMPLEX, 1, (255,120,150), 2)
        
        if cnf > 80:
            cv2.putText(image, 'Hello Pal', (100, 120), cv2.FONT_HERSHEY_COMPLEX, 1, (255,120,150), 2)
            cv2.imshow('FaceId' , image)
           # break
            
        else:
            cv2.putText(image, "sorry I Don't Know you", (250, 450), cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,255), 2)
            cv2.imshow('Face Recognition', image )
    except:
        cv2.putText(image, "No Face Found", (220, 120) , cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,255), 2)
        cv2.putText(image, "Locked", (250, 450), cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,255), 2)
        cv2.imshow('Face Recognition', image )
        pass

    if cv2.waitKey(1) == 13:
        break

cap.release()
cv2.destroyAllWindows()

#ipd.Audio('C://Users//Harry//Documents//MLOPS//bc.wac', autoplay=True )
wb.open('https://www.youtube.com/')
```
